```python
#settings
INPUT_DATA_FOLDER = 'data'
DRIVE_MOUNT_FOLDER = 'drive'
DRIVE_DATA_FOLDER = 'MyDrive/rc_car_tracking/yolo_dataset'
YOLO_TRAINING_WEIGHTS_URL = 'https://pjreddie.com/media/files/darknet53.conv.74'
YOLO_TRAINING_WEIGHTS_FOLDER = 'training_weights'
YOLO_CUSTOM_CFG_FILE_URL = 'https://raw.githubusercontent.com/andreaderuvo/rc_car_tracking/main/datasets/yolo/yolov3_custom.cfg'
```

```python
#imports
import os
import shutil
import urllib.request
```

```python
# Empty folder INPUT_DATA_FOLDER
shutil.rmtree(INPUT_DATA_FOLDER, ignore_errors=True)
os.makedirs(INPUT_DATA_FOLDER, exist_ok=True)
```

```python
#check if the execution environment is Google Colab
try:
  import google.colab
  IN_COLAB = True
except:
  IN_COLAB = False
```

```python
if IN_COLAB:
  from google.colab import drive
  import shutil
  drive.mount(DRIVE_MOUNT_FOLDER, force_remount=True)
  googe_drive_folder = os.path.join(DRIVE_MOUNT_FOLDER, DRIVE_DATA_FOLDER)
  os.makedirs(googe_drive_folder, exist_ok=True)
  for file in os.listdir(googe_drive_folder):
    shutil.copy(os.path.join(googe_drive_folder, file), INPUT_DATA_FOLDER)
  drive.flush_and_unmount()
```

```python
#download custom yolo cfg file
urllib.request.urlretrieve(YOLO_CUSTOM_CFG_FILE_URL, os.path.join(INPUT_DATA_FOLDER, 'yolov3_custom.cfg'))
```

```python
#download darknet53.conv.74: initial yolo weights for training custom data
shutil.rmtree(YOLO_TRAINING_WEIGHTS_FOLDER, ignore_errors=True)
os.makedirs(YOLO_TRAINING_WEIGHTS_FOLDER, exist_ok=True)

urllib.request.urlretrieve(YOLO_TRAINING_WEIGHTS_URL, os.path.join(YOLO_TRAINING_WEIGHTS_FOLDER, 'darknet53.conv.74'))
```
